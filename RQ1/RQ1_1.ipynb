### 2. Run the three models
The three models: BLOOM 7B1, LLaMa 2 7B, LLaMa 3 8B.
Here we run each model with all 1196 prompts in `$DT_PATH/data/toxicity/user_prompts/toxic.jsonl`, with only one response generated for each prompt.

```python
# Runinng BLOOM 7B1
config_path = f'{decoding_trust_path}/src/dt/configs/config.yaml'

config_content = f"""
model_config:
  # model: "hf/mistralai/Mistral-7B-v0.1"
  model: "hf/bigscience/bloom-7b1"
  # model: "hf/meta-llama/Llama-2-13b-chat-hf"
  # model: "hf/meta-llama/Llama-2-7b-chat-hf"
  # model: "hf/meta-llama/Meta-Llama-3-8b"
  type: CHAT
  conv_template: "llama-2"

  model_loader: HF
  torch_dtype: BFLOAT16
  quant_file: null
  # tokenizer_name: mistralai/Mistral-7B-v0.1
  tokenizer_name: bigscience/bloom-7b1
  # tokenizer_name: meta-llama/Llama-2-13b-chat-hf
  # tokenizer_name: meta-llama/Llama-2-7b-chat-hf
  # tokenizer_name: meta-llama/Meta-Llama-3-8b
  trust_remote_code: true
  use_auth_token: true

key: null
dry_run: false

hydra:
  job:
    chdir: false

  callbacks:
    save_job_info:
      _target_: hydra.experimental.callbacks.PickleJobInfoCallback
"""
with open(config_path, 'w') as file:
    file.write(config_content)

!cd $DT_PATH && dt-run +toxicity=realtoxicityprompts-toxic ++toxicity.n=1 ++toxicity.template=1
```
